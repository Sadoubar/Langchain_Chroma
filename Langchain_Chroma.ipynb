```python
# importations des librairies
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_community.embeddings.sentence_transformer import SentenceTransformerEmbeddings
```

```python
# load the document and split it into chunks
# link for data : https://www.dsavocats.com/mailing/Livret_EC_creation_valeur_Environnement.pdf
path = "/content/drive/MyDrive/BNP/Livret_EC_creation_valeur_Environnement.pdf"
loader = PyPDFLoader(path)
documents = loader.load()
```

```python
# document est un liste

print(type(documents))
```

```python
# visualisation du contenu du document index0
documents[0]
```

```python
# split it into chunks (decoupage en chunk)
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
docs = text_splitter.split_documents(documents=documents)
```

```python
print(f"type de l'object docs: {type(docs)}","\n\n")
print(len(docs),"\n\n")
```

```python
# create the open-source embedding function  (choix d'un modele en francais)
model_name = "Sahajtomar/french_semantic"
#model_name = "all-MiniLM-L6-v2"
embedding_function = SentenceTransformerEmbeddings(model_name=model_name)
```

```python
# quuestion
query = "qu'est ce Kenneth Boulding a dit a propos de la croissance"
# save to disk (enregistrement)
db2 = Chroma.from_documents(docs, embedding_function, persist_directory="./chroma_db")
# print results
docs = db2.similarity_search(query,k=5)
print(docs[0].page_content)
```
